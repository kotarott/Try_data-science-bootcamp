### Q21. "2017年12月8日 9時23分16秒"という文字列を、datetimeオブジェクトに変換せよ。

```python
q21 = "2017年12月8日 9時23分16秒"
```

```python
from datetime import datetime as dt

format_q21 = dt.strptime(q21, '%Y年%m月%d日 %H時%M分%S秒')
print(type(format_q21))
```

### Q.22 「2017年12月8日9時23分16秒」から「12371日と23時間23分16秒」戻った日時を求め、"X年Y月Z日"という文字列で表示せよ。

```python
q22 = format_q21

import datetime
time_trip = datetime.timedelta(days=12371, hours=23, minutes=23, seconds=16)

result = q22 - time_trip
print(result.strftime('%Y年%m月%d日'))
```

### Q.24 data/file_handling_assignment/kaken2017.zip内にあるJSONファイルの情報を使って、2017年度に各研究機関が獲得した情報学分野における研究費の合計金額を求めよ。さらに、合計獲得金額の上位20機関を表示せよ。

```python
import os

folder = 'data/file_handling_assignment/kaken2017/kaken2017/'

files = os.listdir(folder)
for file in files:
    print(file)
```

```python
dct = {}
for file in files:
    json_open = open(folder+'/'+file, 'r')
    json_load = json.load(json_open)
    dct[json_load['principal_investigator']['affiliation']] = json_load['budget']

print('合計:', sum(dct.values()))

sorted_dct = sorted(dct.items(), key=lambda x: x[1], reverse=True)

for i in range(0, 20):
    print(sorted_dct[i])
```

### Q25. data/file_handling_assignmentディレクトリにあるbig-cities-pageviews.xmlは、日本語版Wikipediaにある政令指定都市の記事に対するアクセス数（2017/11/29 - 2017/12/04）が記載されている。

```python
from pyquery import PyQuery as pq
```

```python
file_path = 'data/file_handling_assignment/big-cities-pageviews.xml'

with open(file_path, 'r', encoding='utf-8') as f:
    content = f.read()
    doc = pq(content)
    pages = doc('pages')
    print(pages('page[pageid="1642589"]').attr('title'))
```

### Q.26 data/file_handling_assignmentディレクトリにあるbig-cities-pageviews.xmlを読み込み、政令指定都市の記事に対するアクセス数の合計値（2017/11/29 - 2017/12/04）を各都市ごとに表示せよ。

```python
print(pages('page[pageid="1642589"]'))
```

```python
data = dict()
for page in pages('page'):
    element = pq(page)
    city_data = dict()
    for tag in element('pageviews')('pvip'):
        item = pq(tag)
        city_data[item.attr('date')] = int(item.text())
    data[element.attr('title')] = city_data

for key, dct in data.items():
    views = sum(dct.values())
    print(f'{key: <6}\t: {views}')
```
